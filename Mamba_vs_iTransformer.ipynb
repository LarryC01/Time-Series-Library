```python
import re

def extract_log_section(file_path):
    # Regular expression pattern to match the required section
    pattern = re.compile(
        r"(>>>>>>>testing : .+?mse:\d+\.\d+, mae:\d+\.\d+, dtw:-999)",
        re.DOTALL
    )
    
    extracted_sections = []

    with open(file_path, 'r') as file:
        content = file.read()

        # Find all matches in the content
        matches = pattern.findall(content)
        extracted_sections.extend(matches)

    return extracted_sections

def main():
    # Path to the log file
    log_file_path = "scripts/long_term_forecast/Exchange_script/"

    # Extract the sections
    extracted_sections = extract_log_section(log_file_path)

    # Print the extracted sections
    for section in extracted_sections:
        print(section)
        print("=" * 50)  # Separator for clarity

if __name__ == "__main__":
    main()
```
